```python
def get_mongo_client(config, credentials):
    mongo_collection = None
    mongo_client = pymongo.MongoClient(config['mongo_connect_url'].format(credentials['mongo_username'], credentials['mongo_password'], config['mongo_db']))
    mongo_db = mongo_client[config['mongo_db']]
    mongo_collection = mongo_db[config['mongo_collection']]
    try:
        mongo_client = pymongo.MongoClient(config['mongo_connect_url'].format(credentials['mongo_username'], credentials['mongo_password'], config['mongo_db']))
        mongo_db = mongo_client[config['mongo_db']]
        mongo_collection = mongo_db[config['mongo_collection']]
        print('connected...')
    except:
        print('erorr.')
    
    return mongo_collection 
```

```python
mongo_client = get_mongo_client(mongo_db_config, credentials)
```

```python
results = mongo_client.find({})
results_count = results.count(True)
results_count
```

```python
documents = list(results)
```

```python
def get_language(text):
    return detect(text)
```

```python
linkedin_jobs_df['language'] = linkedin_jobs_df.description.progress_apply(get_language)
```

```python
german_states = [x.lower() for x in ["Baden-Württemberg","Bavaria","Berlin","Brandenburg","Bremen","Hamburg","Hesse","Lower Saxony","Mecklenburg-Western","North Rhine-Westphalia","Rhineland-Palatinate","Saarland","Saxony","Saxony-Anhalt","Schleswig-Holstein","Thuringia"]]
```

```python
def location_split(location):
    sp_list = [None] * 3
    if type(location) is not str:
        return sp_list
        
    sp_list = location.split(',')
    sp_list = [x.strip() for x in sp_list]
    while(len(sp_list)!=3):
        sp_list = [None] + sp_list
    if sp_list[1] and sp_list[1].lower().strip() not in german_states:
        sp_list[1] = None
    if sp_list[1] and 'ü' in sp_list[1]:
        sp_list[1] = sp_list[1].replace('ü','ue')
    return sp_list
```

```python
linkedin_jobs_df[['city', 'state', 'country']] = pd.DataFrame([ location_split(x) for x in linkedin_jobs_df['location'].tolist()])
```
